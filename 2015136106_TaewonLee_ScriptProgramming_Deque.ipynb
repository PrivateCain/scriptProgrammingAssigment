```python
class Node:
    def __init__(self, data):
        self.data = data
        self.llink = None
        self.rlink = None

    def __str__(self):
        return str(self.data)

class Deque:
    def __init__(self):
        self.front = None
        self.rear = None

    def __str__(self):
        node = self.front
        print_deque = ' <=> [ '
        while True:
            print_deque += str(node)
            if node == self.rear:
                break
            try: node = node.rlink
            except: break
            print_deque += ', '
        print_deque += ' ] <=>'
        return print_deque
    
    def front(self):
        return self.front.data

    def rear(self):
        return self.front.data

    def empty(self):
        if self.front == self.rear:
            return True
        else:
            return False

    def push_front(self, data):
        new_node = Node(data)
        if self.front == None and self.rear == None:
            self.front = new_node
            self.front.rlink = self.rear
            self.rear = new_node
            self.rear.llink = self.front
        else:
            self.front.llink = new_node
            new_node.rlink = self.front
            self.front = new_node

    def push_back(self, data):
        new_node = Node(data)
        if self.rear == None and self.front == None:
            self.rear = new_node
            self.rear.llink = self.front
            self.front = new_node
            self.front.rlink = self.rear
        else:
            self.rear.rlink = new_node
            new_node.llink = self.rear
            self.rear = new_node

    def pop_front(self):
        if self.empty():
            return
        node = self.front
        value = node.data
        self.front = self.front.rlink
        del node
        return value

    def pop_back(self):
        if self.empty():
            return
        node = self.rear
        value = node.data
        self.rear = self.rear.llink
        del node
        return value
    
    def size(self):
        sizenum=0
        node = self.front
        while(node == self.rear):
            node = node.llink
            sizenum += 1
        return sizenum
```
